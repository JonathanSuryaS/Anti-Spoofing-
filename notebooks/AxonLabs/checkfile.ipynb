```python
import os

dataset_path = r"C:\Users\user\Downloads\AxonLab"

# Count all mp4 files recursively
mp4_files = []
for root, dirs, files in os.walk(dataset_path):
    for f in files:
        if f.endswith('.mp4'):
            mp4_files.append(os.path.join(root, f))

print(f"Total MP4 files: {len(mp4_files)}")
print(f"Total size: {sum(os.path.getsize(f) for f in mp4_files) / 1e9:.2f} GB")

# Show folder structure
for root, dirs, files in os.walk(dataset_path):
    level = root.replace(dataset_path, '').count(os.sep)
    print('  ' * level + os.path.basename(root) + f'/ ({len(files)} files)')
```
